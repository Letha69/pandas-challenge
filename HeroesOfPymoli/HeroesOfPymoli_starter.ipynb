```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
 player_count=(len(purchase_data["SN"].unique()))
     
pd.DataFrame({"Total Players":[player_count]}) # Display the total number of players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate the number of unique items in the DataFrame
item_count =len(purchase_data["Item ID"].unique())

# Calculate the average price for the items
average_price = purchase_data["Price"].mean(axis = 0)

# calculate the number of purchase
number_of_purchase= purchase_data["Item ID"].count()

# Calculate total revenue
total_revenue = purchase_data["Price"].sum()

#Place all the data found into a summary DataFrame
summary_df = pd.DataFrame({"Number of Unique Items":[item_count],
                          "Average Price":average_price,
                          "Number of Purchase":number_of_purchase,
                          "Total Revenue":total_revenue})
# formatting the data
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for aages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
# name the bins
bin_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_data["Total Count"] = pd.cut(purchase_data["Age"], bins, labels=bin_labels)
age1_df = purchase_data[['Total Count','SN']].drop_duplicates(subset = 'SN')

# calculate
age_summary = age1_df.groupby("Total Count").count()
age_counts = age_summary['SN']
age_percentages = (age_counts / 576 * 100)
age_percentages

summary_dataframe3 = pd.DataFrame([age_counts, age_percentages])

summary_data = summary_dataframe3.T

summary_data.columns = ['Total Count', 'Percentage of Players']
#formate
summary_data["Percentage of Players"] = summary_data["Percentage of Players"].astype(float).map("{:,.2f}%".format)

summary_data
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create bins in which to place values based upon Age
bins = [0,9,14,19,24,29,34,39,50]
# Create labels for these bins
bin_labels = ["<10","10-14","15-19","20-24", "25-29", "30-34", "35-39", "40+"]

# Slice and place the data series into a new column inside of the DataFrame
purchase_data["Total Count"] = pd.cut(purchase_data["Age"], bins, labels=bin_labels)

# extract 'Total Count','SN', 'Price'
reduced_row = purchase_data[['Total Count','SN', 'Price']]

reduced_row_grouped = reduced_row.groupby('Total Count')
purchase_counts = reduced_row_grouped['Price'].count()
average_prices = reduced_row_grouped['Price'].mean()
total_spent = reduced_row_grouped['Price'].sum()
purchase_per_person = (total_spent/age_counts)

#Place all the data found into a summary DataFrame
summary_spend_df = pd.DataFrame({"Purchase Count":purchase_counts,
                          "Average Purchase Price":average_prices,
                          "Total Purchase Value":total_spent,
                          "Avg Total Purchase per Person":purchase_per_person})
summary_spend_df["Average Purchase Price"] = summary_spend_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
summary_spend_df["Total Purchase Value"] = summary_spend_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
summary_spend_df["Avg Total Purchase per Person"] = summary_spend_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
summary_spend_df
```
